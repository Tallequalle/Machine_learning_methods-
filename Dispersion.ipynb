```python
import pandas as pd
import seaborn as sns

def Dispersion():
    data = pd.read_excel('../DATA.xls')
    print(type(data))
    clear_df = data.dropna()
    dispersion = []
    lenght = len(clear_df)
    for i in data.columns:
        data_series = pd.Series(clear_df[i])
        sum = 0
        for j in data_series:
            difference = j - data_series.mean()
            sum = sum + ((difference)**2)
        dispersion.append(sum)
    dispersion = [element/(lenght - 1) for element in dispersion]
    print("The sum of the differences in the square:", dispersion)
    index = dispersion.index(min(dispersion))
    print("The most stable indicator:", clear_df.columns[index])
    barplot(clear_df.columns,dispersion)

def barplot(x, y):
    sns.set(rc={'figure.figsize':(15,9)})
    sns.barplot(x = x, y = y)
    
Dispersion()
```
