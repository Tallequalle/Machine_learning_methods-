```python
import numpy as np
import math
```

```python
def gradient_descent(alpha = 0.05,iter = 100):
    theta0 = []
    theta1 = []
    cost = []
    x = np.array([1,4,7,5,3,6,9,8,10,2,0])
    y = np.array([0,0,1,1,0,1,1,1,1,0,0])
    x = np.column_stack((np.ones(len(x)),x))
    theta = np.zeros(2)
    for i in range(iter):
        pred = predict(x,theta)
        print("Pred", pred)
        t0 = theta[0] - alpha *(pred - y).mean()
        t1 = theta[1] - alpha *((pred - y)* x[:,1]).mean()
        theta = np.array([t0,t1])
        cst = calculateCost(x,theta,y)
        theta0.append(t0)
        theta1.append(t1)
        cost.append(cst)
        if i%10==0:
            print(f"Iteration: {i+1},Cost = {cst},theta = {theta}")
    x = [int(input("Input the number[0,10]: "))]
    x = np.column_stack((np.ones(len(x)),x))
    print("Predict:",predict(x,theta))
```

```python
def predict(x,theta):
    expon_lst = []
    hypothesis_lst = []
    expon = -np.dot(x,theta)
    for i in expon:
        i = math.exp(i)
        expon_lst.append(i)
    for i in expon_lst:
        i = 1/(1 + i)
        hypothesis_lst.append(i)
    for i in range(len(hypothesis_lst)):
        if hypothesis_lst[i] >= 0.5: hypothesis_lst[i] = 1
        else: hypothesis_lst[i] = 0
    return hypothesis_lst
```

```python
def calculateCost(x,theta,y):
    pred = predict(x,theta)
    cost_lst = []
    for i in range(len(pred)):
        if pred[i] - y[i] == 1:
            cost_lst.append(1)
        else:
            cost_lst.append(0)
    return sum(cost_lst)/len(cost_lst)
```
